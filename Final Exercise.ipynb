```python
import random

random_choices = { 0 : "rock", 1 : "paper", 2 : "scissors" }

print("Rock, Paper & Scissors Game")

while True :
    user_input = input("Would you like to play a game of \"Rock, Paper & Scissors\" [yes/no]? ").lower()
    if user_input != "no" and user_input != "yes" :
        print("I do not understand your request, please try again.")
        continue
    if user_input.lower() == "no" :
        print("Goodbye.")
        break
    user_input = input("Please enter a choice of \"rock\", \"paper\" or \"scissors\": ").lower()
    if user_input != "rock" and user_input != "paper" and user_input != "scissors" :
        print("Invalid choice. Your valid choices are \"rock\", \"paper\" or \"scissors\".")
        continue
    choice = random.randint(0,2)
    comp = random_choices[choice]
    # print(comp)
    print(f"Computer chose {comp}, you chose {user_input}.")
    if ( comp == "rock" and user_input == "scissors" ) or ( comp == "scissors" and user_input == "paper" ) or ( comp == "paper" and user_input == "rock" ) :
        print("You lose.")
        print(f"{comp.capitalize()} beats {user_input}")
    elif ( comp == "scissors" and user_input == "rock" ) or ( comp == "paper" and user_input == "scissors" ) or ( comp == "rock" and user_input == "paper" ) :
        print("You win!")      
        print(f"{user_input.capitalize()} beats {comp}")
    else :
        print("It's a draw.")
```

```python
shoppingcart = {}

menu = {
    "Coffee" : {
        "Espresso" : 1.50,
        "Americano" : 2.00,
        "Caffe Latte" : 2.25,
        "Mocha" : 2.25,
        "Cappucino" : 2.50,
        "Macchiato" : 2.50,
    },
    "Tea" : {
        "White Tea" : 1.50,
        "Green Tea" : 2.00,
        "Oolong Tea" : 2.25,
        "Black Tea" : 2.25,
        "Earl Grey" : 2.50,
        "Matcha" : 2.50,
    },
    "Pastries" : {
        "Croissant" : 5.50,
        "Cinnamon Roll" : 4.50,
        "Banana Bread" : 4.25,
        "Pumpkin Bread" : 3.75,
        "Sticky Bun" : 4.25,
        "Apple Turnover" : 3.75,
    },
}

tax_rate = 0.06

def display_menu() :
    for category in menu :
        print(f"=== [ {category} ] ===")
        for item, value in menu[category].items() :
            print(f" * {item:<15} - ${value:,.2f}")
        print()
        
def search_menu(item) :
    for category in menu :
        if item in menu[category] :
            return menu[category][item]
    return 0

def print_receipt() :
    subtotal = 0
    print("====================================")
    print("MC DataWork Cafe!")
    print("------------------------------------")
    for item in shoppingcart :
        price = search_menu(item)
        quantity = shoppingcart[item]
        print(f"{item:<15} ... {quantity:3} x ${price:,.2f} = ${quantity*price:>5,.2f}")
        subtotal += quantity * price
    
    tax = subtotal * tax_rate
    print("====================================")
    print(f"Subtotal:      ${subtotal:>5.2f}")
    print(f"Tax:           ${tax:>5.2f}")
    print(f"Total:         ${subtotal + tax:>5.2f}")
    print("====================================")
        
print("Welcome to MC DataWork Cafe!", end="\n\n")
while True :
    user_input = input("Would you like to see the menu? (yes/no) ").lower()
    if user_input != "yes" and user_input != "no" :
        print("I am sorry, I did not understand your answer.")
        continue
    if user_input == "yes" :
        display_menu()
    user_input = input("Would you like to order? (yes/no) ").lower()
    if user_input != "yes" and user_input != "no" :
        print("I am sorry, I did not understand your answer.")
        continue
    if user_input == "no" :
        if len(shoppingcart) > 0 :
            print_receipt()

        print("\n\nThank you for visiting MC DataWork Cafe!")
        break
    else :
        user_input = input("(Please type in the name exactly) What did you want to order? ")
        if not search_menu(user_input) :
            print(f"I am sorry, I am not able to find {user_input}. Please try again.")
            continue
        quantity = input("How many did you want to order? ")
        try :
            quantity = int(quantity)
        except :
            print("Please enter a numeric value.")
        if quantity <= 0 :
            print("I am sorry, you cannot purchase a negative quantity.")
            continue
        shoppingcart[user_input] = shoppingcart.get(user_input, 0) + quantity
        # print(shoppingcart)
```
